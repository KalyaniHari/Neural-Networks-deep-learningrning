1. Data Manipulation

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Upload the file manually in Google Colab
from google.colab import files
uploaded = files.upload()

# Read the CSV file
df = pd.read_csv('data.csv')

# (b) Show the basic statistical description about the data
print("Basic Statistical Description:\n", df.describe())

# (d) Check if the data has null values
print("\nNull Values Count:\n", df.isnull().sum())

# (i) Replace null values with the mean
df.fillna(df.mean(), inplace=True)
print("\nNull values after replacement:\n", df.isnull().sum())

# (e) Select two columns (Pulse and Calories) and aggregate the data
aggregations = df[['Pulse', 'Calories']].agg(['min', 'max', 'count', 'mean'])
print("\nAggregated Data (Pulse & Calories):\n", aggregations)

# (f) Filter rows with Calories between 500 and 1000
calories_filtered_500_1000 = df[(df['Calories'] >= 500) & (df['Calories'] <= 1000)]
print("\nFiltered Data (Calories between 500-1000):\n", calories_filtered_500_1000)

# (g) Filter rows with Calories > 500 and Pulse < 100
calories_pulse_filtered = df[(df['Calories'] > 500) & (df['Pulse'] < 100)]
print("\nFiltered Data (Calories > 500 & Pulse < 100):\n", calories_pulse_filtered)

# (h) Create a new dataframe without the "Maxpulse" column
df_modified = df.drop(columns=['Maxpulse'])
print("\nNew dataframe (Without Maxpulse Column):\n", df_modified.head())

# (i) Delete the "Maxpulse" column from the original dataframe
df.drop(columns=['Maxpulse'], inplace=True)
print("\nUpdated DataFrame after deleting 'Maxpulse':\n", df.head())

# (j) Convert the datatype of Calories column to integer
df['Calories'] = df['Calories'].astype(int)
print("\nDataframe after converting 'Calories' to int:\n", df.head())

# (k) Create a scatter plot for Duration and Calories
plt.figure(figsize=(8, 5))
plt.scatter(df['Duration'], df['Calories'], color='blue', alpha=0.5)
plt.xlabel("Duration")
plt.ylabel("Calories")
plt.title("Scatter Plot: Duration vs Calories")
plt.grid(True)
plt.show()
```

2. Linear Regression

```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Load the dataset (assuming file is uploaded manually in Colab)
from google.colab import files
uploaded = files.upload()

# Read the uploaded file
salary_data = pd.read_csv("Salary_Data.csv")

# (b) Split the data into train and test sets (1/3 reserved for testing)
X = salary_data[['YearsExperience']]  # Features
y = salary_data['Salary']             # Target variable

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=42)

# (c) Train the model and make predictions
model = LinearRegression()
model.fit(X_train, y_train)  # Train the model

y_pred = model.predict(X_test)  # Predictions on test data

# (d) Calculate the Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")

# (e) Visualize train and test data
plt.figure(figsize=(10, 6))

# Scatter plot for training data
plt.scatter(X_train, y_train, color='blue', label='Training Data')

# Scatter plot for testing data
plt.scatter(X_test, y_test, color='green', label='Testing Data')

# Regression line
plt.plot(X_train, model.predict(X_train), color='red', label='Regression Line')

plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Train and Test Data with Regression Line')
plt.legend()
plt.grid(True)
plt.show()
```
